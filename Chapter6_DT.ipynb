```
from sklearn.datasets import make_moons
Xm, ym = make_moons(n_samples=1000, noise=0.4, random_state=42)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xm,ym,test_size=0.2)
```

```
from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier()
```

```
from sklearn.model_selection import GridSearchCV
param_grid=[{'max_leaf_nodes':[5,6,7,8,9,10],
             'min_samples_leaf':[1,2,3],
             'max_depth':[3,4,5]}]
grid=GridSearchCV(dt_clf,param_grid,cv=5,verbose=2)
```

```
grid.fit(X_train,y_train)
```

```
print(grid.best_score_)
grid.best_estimator_
```

```
dt_best=DecisionTreeClassifier(max_leaf_nodes=5,max_depth=3)
dt_best.fit(X_train,y_train)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_true=y_test,y_pred=dt_best.predict(X_test))
```

```
from sklearn.model_selection import ShuffleSplit
rs=ShuffleSplit(1000,train_size=100,random_state=42)
rs.get_n_splits(X_train)
```

```
X_subs=[]
y_subs=[]
for train_index in rs.split(X_train):
  X_subs.append(X_train[train_index[0]])
  y_subs.append(y_train[train_index[0]])
```

```
from sklearn.base import clone
dts=[]
for i in range(1000):
  dt=clone(dt_best).fit(X_subs[i],y_subs[i])
  dts.append(dt)
```

```
from scipy.stats import mode
pred=[]

predictions=[]
for i in range(1000):
  predictions.append(dts[i].predict(X_test))

modes,n=mode(predictions)
modes[0]
```

```
accuracy_score(y_pred=modes.reshape([-1]),y_true=y_test)
```
